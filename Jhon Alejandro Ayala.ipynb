$\textbf{Primer Punto}$

Tenemos un problema de la forma $Bx=y$, en donde $B=A^{T}A$ una matriz cuadrada y $y=A^{T}b$. Y nuestro objetivo es encontrar las condiciones sobre $A$ y $b$ tal que el problema planteado tenga solucion.

Sabemos que el problema $Bx=y$ tiene solución unica si y solo si $\text{det}(B)\neq0$, entonces nuestro problema se resume en encontrar condiciones sobre $A$ de manera que $A^{T}A$ tenga determinante diferente de 0.

Tenemos que si $A$ es cuadrada entonces $\text{det}(A^{T}A)=\text{det}^2(A)$ y bastaria con que $A$ sea invertible para que el problema tenga solucion unica.

Ahora, supongamos que el numero de filas de $A$ es mucho mayor al numero de columnas, es decir, tomamos $A_{m\times n}$ con $m>>n$, es decir, un sistema sobre condicionado, que es lo que buscamos. 

Por un lado vemos que, dado $x\in\mathbb{R}^n$ tenemos

$x^{T}A^{T}Ax=(Ax)^{T}Ax=||Ax||^2\geq0$

Por lo tanto tenemos que $A^{T}A$ es simetrica semidefinida positiva

Tomamos

$$A=[x_1\phantom{aa}x_2\phantom{aa}\dots\phantom{aa} x_n]$$

con $x_i\in\mathbb{R}^m$. Entonces 

$A^{T}A=\begin{equation}
\begin{pmatrix}
||x_1||^2 & x_1\cdot x_2 & \cdots & x_1\cdot x_n\\
x_1\cdot x_2 & ||x_2||^2 & \cdots & x_2\cdot x_n\\
\vdots & \vdots & \ddots & \vdots\\
x_1\cdot x_n & x_2\cdot x_n & \cdots & ||x_n||^2
\end{pmatrix}
\end{equation}$


Primero buscamos que no haya una fila de ceros para que el determinante sea no nulo (que es equivalente a teer una columna de ceros pues $A^{T}A$ es simetrica), esto lo podemos evitar si logramos que la diagonal sea no nula, y como tenemos que $||x_i||^2=0$ si y solo si $x_i=0$ entonces nuestra primera condicion es:

* $A$ no debe tener una columna de ceros

Ahora, dado que la diagonal es no nula, nos basta ver que las filas de $A^{T}A$ son linealmente independientes (o columnas ya que es simetrica).

Vemos que para que las filas de $A^{T}A$ son linealmente dependientes debe ocurrir que 

$x_i\cdot x_i=c_j(x_i\cdot x_j)+c_k(x_i\cdot x_k)$

Y dado que la matriz es simetrica tambien tenemos que 

$x_i\cdot x_j=c_j(x_j\cdot x_j)+c_k(x_j\cdot x_k)$

$x_i\cdot x_k=c_j(x_k\cdot x_j)+c_k(x_k\cdot x_k)$

Y esto solo puede ocurrir si

$x_i=c_jx_j+c_kx_k$

Entonces las filas (y columnas) de $A^{T}A$ son linealmente dependientes si y solo si las columnas de $A$ son linealmente dependientes. Esto nos da otra condicion sobre $A$. 

* Las columnas de $A$ deben ser linealmente independientes

Con esto vemos que si $A$ tiene mas filas que columnas, no se puede encontrar solucion por metodo de minimos cuadrados pues las columnas serian linealmente dependientes.

Ahora, vemos que la segunda condicion contiene a la primera, por lo tanto, hemos demostrado que para que el problema $A^{T}Ax=A^{T}b$ tenga solucion unica entonces las columnas de $A$ deben ser linealmente independientes.
